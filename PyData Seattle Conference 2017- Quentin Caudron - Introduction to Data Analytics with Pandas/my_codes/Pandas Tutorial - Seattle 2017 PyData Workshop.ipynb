```python
people = pd.read_csv("data/department_members.csv", index_col='date',parse_dates=True)
```

```python
people.head()
```

```python
#inner join - no overlap in two datasets, delete unmatched,
#which is the default for pandas
#outer join - keep bothd
daily_data = daily_data.join(people, how='outer').interpolate(method='nearest')
```

```python
daily_data['coffees_per_person'] = daily_data.coffees_made_today / daily_data.members
    
daily_data.head(n=10)
```

```python
daily_data.coffees_per_person.plot()
```
